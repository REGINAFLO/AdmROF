PRACTCA 2 REGINA ORTEGA FLORES

El problema que un supermercado quiere tener el control de las ventas por productos de nuevo lanzamiento , en el cual necesitan saber el monto total que vendio en un día , Por tal motivo los datos que se necesitan son ID (int),Nombre del cliente(string) , producto(string) , precio (double), cantidad (int) y tototal (double). Los int=4 , double=8 y string=1037.
Para poder calcular el tamaño de la base de datos se necesita saber cuanto registros aproximadamente se tendran(en nuestro caso 30000 registros) y el numero de filas por pagina que se desean-(en nuestro ejemplo 20 filas por pagina) ya que con estos dos datos calcularemos el número de paginas que se necesitan para la base de datos(1500) = #registros (30000) / #filas por pagina(20).
La forma de obtener el tamaño de la base de datos es mediante el siguiente procedimiento:
1.-Calcular tamaño de  1 fila = suma de bytes de cada atributo.
2.-Calculamos tamaño de 1 pagina = numero de filas por pagina *tamaño de una fila(lo obtuvimos en el primer paso)
3.- Calculamos tamaño de todas paginas= # de paginas que se necesitan para la base de datos *tamaño de 1 pagina ( la obtuvimos en el paso 2).
